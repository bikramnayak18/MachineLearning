```python
def computeCost(X, y, theta):
    m,n = X.shape
    h = np.dot(X,theta)
    delta=0
    for i in range(m):
        delta += (h[i]-y[i])**2     
    cost = 1/(2*m)*delta
    return cost
```

```python
m,n = X.shape
ones = np.ones([m,1])
inputX = np.concatenate((ones,X), axis=1)

computeCost(inputX, y, theta)
```

```python
def gradientDescent(X,y, theta,alpha, iteration):
    m,n = X.shape
    h = np.dot(X,theta)
    delta = np.zeros((m,1))
    cost = np.zeros((iteration,1))
    thetaArr = np.zeros((iteration,2))
    for i in range(iteration):
        pred = np.dot(X,theta) 
        for j in range(m):
            delta[j] = (pred[j]-y[j])  
        theta = theta - (alpha /m) * np.dot(X.T,delta)
        cost[i] = computeCost(inputX, y, theta)
        thetaArr[i] = theta.T
    return theta,cost,thetaArr
```

```python
X = data.iloc[:,:-1].values
y= data.iloc[:,1].values
(m,n) = X.shape
theta =np.zeros((n+1,1))
ones = np.ones([m,1])
inputX = np.concatenate((ones,X), axis=1)
theta,cost,thetaArr = gradientDescent(inputX, y, theta, 0.01,1500)
```
